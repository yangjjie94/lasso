```python
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import os
import glob

%matplotlib inline
```

```python
os.chdir("..")
```

```python
# Read in the image
paths = glob.glob('data/600V350Q401/*.jpg')
step = 1
for index, f1, f2 in zip(range(len(paths)-1), paths[:-1:step], paths[1::step]):
    if index > step:
        break
    image1 = cv.imread(f1)
    image1 = cv.cvtColor(image1, cv.COLOR_BGR2RGB)
    image1_gray = cv.cvtColor(image1, cv.COLOR_RGB2GRAY)
    image2 = cv.imread(f2)
    image2 = cv.cvtColor(image2, cv.COLOR_BGR2RGB)
    image2_gray = cv.cvtColor(image2, cv.COLOR_RGB2GRAY)
    delta_image = image2 - image1
    delta_gray_image = image2_gray - image1_gray

plt.imshow(delta_gray_image, cmap='gray')
```

```python
cv.imwrite('timd_d_gray.jpg',delta_gray_image)
```

```python
fig = plt.figure()

# def f(x, y):
#     return np.sin(x) + np.cos(y)

# x = np.linspace(0, 2 * np.pi, 120)
# y = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)
# ims is a list of lists, each row is a list of artists to draw in the
# current frame; here we are just animating one artist, the image, in
# each frame
ims = []
paths = glob.glob('data/600V350Q401/*.jpg')
step = 1
for index, f1, f2 in zip(range(len(paths)-1), 
                         paths[0:-1:step], 
                         paths[1: :step]):
    if index > step + 100:
        break
    image1 = cv.imread(f1)
    image1 = cv.cvtColor(image1, cv.COLOR_BGR2RGB)
    image1_gray = cv.cvtColor(image1, cv.COLOR_RGB2GRAY)
    image2 = cv.imread(f2)
    image2 = cv.cvtColor(image2, cv.COLOR_BGR2RGB)
    image2_gray = cv.cvtColor(image2, cv.COLOR_RGB2GRAY)
    delta_image = image2 - image1
    delta_gray_image = image2_gray - image1_gray
    plt.imshow(delta_gray_image, cmap='gray')
# for i in range(60):
#     x += np.pi / 15.
#     y += np.pi / 20.
#     im = plt.imshow(f(x, y), animated=True)
#     ims.append([im])

ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                repeat_delay=1000)

# ani.save('dynamic_images.mp4')

plt.show()
```
